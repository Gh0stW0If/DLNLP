```python
#导入相关包
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import string
import math
import nltk.util
import gc
```

```python
#读取英文语料库
# english_text = gutenberg.raw('milton-paradise.txt')
english_text = gutenberg.raw('shakespeare-hamlet.txt')
#转化为小写
english_text = english_text.lower()

#去掉标点符号
english_text = english_text.translate(str.maketrans('', '', string.punctuation))
#tokenize
english_tokens_word = word_tokenize(english_text)
#去掉stopwords
english_tokens_word = [word for word in english_tokens_word if word not in stopwords.words('english')]
english_tokens_char = list(english_text)
print(english_tokens_word)
```

```python
def cal_entropy(tokens):
    bigram = list(nltk.bigrams(tokens))
    
    total_uni = len(tokens)
    total_bi = len(bigram)
    
    print("一阶总词数：", total_uni)
    print("二阶总词数：", total_bi)
    
    unigram_freq = nltk.FreqDist(tokens)
    bigram_freq =  nltk.FreqDist(bigram)
    
    unigram_entropy = 0
    bigram_entropy = 0
    trigram_entropy = 0

    for freq in unigram_freq.values():
        prob = freq/total_uni
        unigram_entropy -= prob * math.log2(prob)
    print("一阶熵：", unigram_entropy)
    for bigram,freq in bigram_freq.items():
        prob = freq/total_bi
        p_condirion = freq/unigram_freq[bigram[0]]
        bigram_entropy -= prob * math.log2(p_condirion)
    print("二阶熵：", bigram_entropy)
    trigram = list(nltk.trigrams(tokens))
    total_tri = len(trigram)
    print("三阶总词数：", total_tri)
    trigram_freq = nltk.FreqDist(trigram)
    del tokens
    del bigram
    gc.collect() 
    for trigram,freq in trigram_freq.items():
        prob = freq/total_tri
        p_condirion = freq/bigram_freq[trigram[:2]]
        trigram_entropy -= prob * math.log2(p_condirion)
    print("三阶熵：", trigram_entropy)
```

```python
def cal_mean_entropy(tokens):
    #计算一阶平均熵
    total_uni = len(tokens)
    unigram_freq = nltk.FreqDist(tokens)
    unigram_entropy = 0
    for freq in unigram_freq.values():
        prob = freq/total_uni
        unigram_entropy -= math.log2(prob)/total_uni
    print("一阶平均熵：", unigram_entropy)
```

```python
cal_entropy(english_tokens_word)
cal_entropy(english_tokens_char)
cal_mean_entropy(english_tokens_word)
cal_mean_entropy(english_tokens_char)
```

```python
#加载中文语料库
import jieba
import json

# 读取文件
wiki_data = []
parent_path = r'D:\homework2025\DLandNLP\DL-nlp2025-main\wiki_zh'
#读取目录下的所有文件
import os
for root, dirs, files in os.walk(parent_path):
    for file in files:
        file_path = os.path.join(root, file)
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                # 解析每行的 JSON 数据
                data = json.loads(line)
                wiki_data.append(data)



content = []
for i in range(len(wiki_data)):

    content.append(wiki_data[i]['text'])

#拼接
# print(content[3])
print(len(content))
```

```python
import sys
import gc
m3 = sys.getsizeof(content) / 1024 /1024
m4 = sys.getsizeof(wiki_data) / 1024 /1024
print(m3)
print(m4)
del wiki_data
gc.collect()
```

```python
#预处理
import opencc
import re
# 繁體轉簡體
rule = re.compile(r"[^a-zA-Z0-9\u4e00-\u9fa5]")
converter = opencc.OpenCC('t2s.json')
stopwords = [line.strip() for line in open('D:\homework2025\DLandNLP\DL-nlp2025-main\cn_stopwords.txt', 'r', encoding='utf-8').readlines()]
words = []
for i in range(len(content)):
    content[i] = converter.convert(content[i])
    content[i] = rule.sub('', content[i])
    content[i] = re.sub('[a-zA-Z]','',content[i])
    temp = jieba.lcut(content[i])
    temp2 = [word for word in temp if word not in stopwords]
    words.extend(temp2)
    temp.clear()
    temp2.clear()

print(sys.getsizeof(words) / 1024 /1024)
del content
gc.collect()
print(sys.getsizeof(converter) / 1024 /1024)
```
