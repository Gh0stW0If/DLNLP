```python
import os
import jieba
import re
import random
import opencc
```

```python
#加载语料库和预处理分词
#遍历语料文件夹
texts_folder = r"D:\homework2025\DLandNLP\texts"
ind_file = r"D:\homework2025\DLandNLP\texts\inf.txt"
files = []
with open(ind_file, 'r', encoding='ANSI') as f:
    # 读取文件列表并加上.txt后缀
    line = f.readline()
    #用,分割
    files = line.split(',')


print(files)
texts = []
for file in [file.strip() + '.txt' for file in files]:
    #读取文件内容
    with open(os.path.join(texts_folder, file), 'r', encoding='ANSI') as f:
        text = f.read()
        texts.append(text)
print(len(texts))
#预处理
#删除开头和结尾的‘本书来自www.cr173.com免费txt小说下载站 更多更新免费电子书请关注www.cr173.com’
for i in range(len(texts)):
    texts[i] = texts[i].replace("本书来自www.cr173.com免费txt小说下载站\n更多更新免费电子书请关注www.cr173.com", "")

for i in range(len(texts)):

    #删除空白符
    texts[i] = texts[i].replace("\u3000", "")
def fix_chinese_text(text):
    # 移除句中换行（不在句号、问号、叹号后）
    text = re.sub(r'(?<![。！？])\n(?!\n)', '', text)  
    # 统一段落换行（确保段落间保留空行）
    text = re.sub(r'\n{2,}', '\n\n', text)  
    return text

for i in range(len(texts)):
    texts[i] = fix_chinese_text(texts[i])
#先分段
paragraphs = []
for text in texts:
    # 按照两个换行符分割文本
    paragraph = text.split('\n')
    paragraphs.extend(paragraph)

# print(paragraphs[:10])  # 打印前10段文本
#分词

rule = re.compile(r"[^a-zA-Z0-9\u4e00-\u9fa5]")
converter = opencc.OpenCC('t2s.json')
stopwords = [line.strip() for line in open('D:\homework2025\DLandNLP\DL-nlp2025-main\cn_stopwords.txt', 'r', encoding='utf-8').readlines()]
#预处理
def preprocess_text(text,keep_stopwords=False):
    #去除标点符号和英文
    text = rule.sub('', text)
    text = re.sub('[a-zA-Z]','', text)
    #去除数字
    text = re.sub(r'\d+', '', text)
    #简体转繁体
    text = converter.convert(text)
    #分词
    words = list(jieba.cut(text))
    #去除停用词
    if not keep_stopwords:
        words = [word for word in words if word not in stopwords]
    return words


all_words = []
for text in paragraphs:
    words = preprocess_text(text,keep_stopwords=True)#保留停用词
    # print(words[:10])  # 打印前10个分词
    #去除空字符串
    words = [word for word in words if word !='']
    #去除空段落
    if len(words) < 1:
        continue
    all_words.append(words)

print(all_words[:100])
```

```python
#使用word2vec训练模型
from gensim.models import Word2Vec
#设置随机种子
random.seed(42)
#训练模型
model = Word2Vec(sentences=all_words, vector_size=20, window=5, min_count=3, workers=4, sg=0,epochs=10,negative=5)
#输出令狐冲的词向量
word_vector = model.wv['令狐冲']
print(word_vector)
#输出和令狐冲最相关的10个词
similar_words = model.wv.most_similar('令狐冲', topn=10)
print(similar_words)
#保存模型
model.save(r'D:\homework2025\DLandNLP\code\hw3\word2vec.model')
```

```python
#使用glove训练模型

# 构建空的词表`
from mittens import GloVe

os.environ['OMP_NUM_THREADS'] = '4'
#转成numpy数组
import numpy as np

#将all_words计算成词频矩阵
from sklearn.feature_extraction.text import CountVectorizer

#将all_words转成字符串
all_words_str = [' '.join(words) for words in all_words]
from collections import defaultdict
import numpy as np

def build_cooccurrence(sentences, vocab, window_size=2):
    word_to_id = {word: idx for idx, word in enumerate(vocab)}
    cooccurrence = defaultdict(float)

    for sentence in sentences:
        words = [w for w in sentence.split() if w in word_to_id]
        for i, center_word in enumerate(words):
            center_id = word_to_id[center_word]
            # Look at context window
            for j in range(max(0, i - window_size), min(len(words), i + window_size + 1)):
                if i == j:
                    continue  # skip center word
                context_word = words[j]
                context_id = word_to_id[context_word]
                # Weight by distance (optional, like in original GloVe paper)
                distance = abs(i - j)
                cooccurrence[(center_id, context_id)] += 1.0 / distance
    return cooccurrence, word_to_id

def cooc_dict_to_matrix(cooccurrence, vocab_size):
    cooc_matrix = np.zeros((vocab_size, vocab_size))
    for (i, j), value in cooccurrence.items():
        cooc_matrix[i, j] = value
    return cooc_matrix
```

```python
# Build vocab from most common words
from collections import Counter
#从all_words中提取出所有的单词
sentences = [" ".join(words) for words in all_words]
flat_words = " ".join(sentences).split()
most_common = [w for w, _ in Counter(flat_words).most_common(5000)]
vocab = list(set(most_common))

# Build cooccurrence
cooc_dict, word_to_id = build_cooccurrence(sentences, vocab, window_size=5)
cooc_matrix = cooc_dict_to_matrix(cooc_dict, vocab_size=len(vocab))
```

```python
from mittens import Mittens,GloVe

mittens_model = GloVe(n=20,max_iter=10000,display_progress=100
)
#训练模型
result =mittens_model.fit(cooc_matrix)
#输出fox的词向量
 #找到fox的索引
word_index = word_to_id['令狐冲']
#输出词向量
print(result.shape)
print(result[word_index])
#输出和fox最相似的10个词
similar_words = np.argsort(result[word_index])[-10:]
idx_to_word = {idx: word for word, idx in word_to_id.items()}
for idx in similar_words:
    print(idx_to_word[idx], result[word_index][idx])
```
